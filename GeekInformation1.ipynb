```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/GeekInfo/E-Commerce sales .csv",encoding='unicode_escape')
df
```

```python
#checking for missing values in all the columns
df.isnull().sum()
```

```python
#df.drop(columns=['Description','CustomerID'],axis=1,inplace=True)
#df
df.dropna()
```

```python
df.describe()
```

```python
#Cost and the quantity cannot be negative
df = df[df['Quantity'] >= 0]
df = df[df['UnitPrice'] >= 0]
df
```

```python
df.describe()
df
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df
```

```python
df.boxplot()
```

```python
#removing outliers(removing rare orders)
col_1 = 'Quantity'
col_2 = 'UnitPrice'
Q1 = df[col_1].quantile(0.25)
Q3 = df[col_1].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_mask = (df[col_1] < lower_bound) | (df[col_1] > upper_bound)
df_new = df[~outliers_mask]

Q1 = df[col_2].quantile(0.25)
Q3 = df[col_2].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_mask = (df[col_2] < lower_bound) | (df[col_2] > upper_bound)
df_new = df[~outliers_mask]

df_new
```

```python
df_new = df_new[df_new['Quantity']<10000]
# Scatter plot of Quantity vs UnitPrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UnitPrice', y='Quantity', data=df_new)
plt.title('Scatter plot of Quantity vs UnitPrice')
plt.show()
```

```python
df_2 = df.copy()
df_2.set_index('InvoiceDate', inplace=True)
monthly_sales = df_2.resample('M').sum()
```

```python
monthly_sales[['Quantity','UnitPrice']]
```

```python
# RFM Analysis
recency = df_new['InvoiceDate'].max() - df_new.groupby('CustomerID')['InvoiceDate'].max()
frequency = df_new.groupby('CustomerID').size()
monetary_value = df_new.groupby('CustomerID')['UnitPrice'].sum()

rfm_df = pd.DataFrame({
    'Recency': recency.dt.days,
    'Frequency': frequency,
    'MonetaryValue': monetary_value
})

# Clustering (using K-means)
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_df)

kmeans = KMeans(n_clusters=3, random_state=42)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming 'df_new' is your DataFrame
# Assuming 'rfm_df' has already been created with clustering results

# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot for each cluster
for cluster_label in range(3):
    cluster_data = rfm_df[rfm_df['Cluster'] == cluster_label]
    ax.scatter(cluster_data['Recency'], cluster_data['Frequency'], cluster_data['MonetaryValue'], label=f'Cluster {cluster_label}')

# Set labels and title
ax.set_xlabel('Recency')
ax.set_ylabel('Frequency')
ax.set_zlabel('MonetaryValue')
ax.set_title('K-means Clustering of RFM')

# Show the legend
ax.legend()

# Show the plot
plt.show()
```

```python
# Product Performance Analysis
top_products = df.groupby('Description')['Quantity'].sum().sort_values(ascending=False).head(10)

# Sales Trends
monthly_sales['Revenue'] = monthly_sales['Quantity'] * monthly_sales['UnitPrice']
```

```python
# Import necessary libraries
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
from math import sqrt

# Extract relevant time series data
time_series_data = df.groupby('YearMonth')['Quantity'].sum()

# Model Training
order = (1, 1, 1)  # Example order, you may need to fine-tune
seasonal_order = (1, 1, 1, 12)  # Example seasonal order, you may need to fine-tune
sarima_model = SARIMAX(time_series_data, order=order, seasonal_order=seasonal_order)
sarima_result = sarima_model.fit(disp=False)

# Forecasting
sarima_forecast = sarima_result.get_forecast(steps=12)
sarima_predictions = sarima_forecast.predicted_mean

# Evaluate the model
rmse_sarima = sqrt(mean_squared_error(time_series_data[-12:], sarima_predictions))
print(f"RMSE for SARIMA: {rmse_sarima}")


# Import necessary libraries
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error

# Feature Engineering
df['Timestamp'] = pd.to_datetime(df['InvoiceDate'])
df.set_index('Timestamp', inplace=True)

# Create lag features
for i in range(1, 13):
    df[f'lag_{i}'] = df['Quantity'].shift(i)

# Drop missing values created by lag
df.dropna(inplace=True)

# Split data into training and testing sets
train_size = int(len(df) * 0.8)
train, test = df[0:train_size], df[train_size:]

# Define features and target variable
features = [col for col in df.columns if col.startswith('lag_')]
X_train, y_train = train[features], train['Quantity']
X_test, y_test = test[features], test['Quantity']

# Model Training
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Forecasting
rf_predictions = rf_model.predict(X_test)

# Evaluate the model
rmse_rf = sqrt(mean_squared_error(y_test, rf_predictions))
print(f"RMSE for RandomForest: {rmse_rf}")
```

```python
# Convert Period index to datetime
time_series_data.index = time_series_data.index.to_timestamp()

# Plotting SARIMA forecast
plt.figure(figsize=(12, 6))
plt.plot(time_series_data.index, time_series_data, label='Actual')
plt.plot(sarima_predictions.index, sarima_predictions.values, color='red', label='SARIMA Forecast')
plt.title('SARIMA Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
# Plotting RandomForest forecast
plt.figure(figsize=(12, 6))
plt.plot(test.index, y_test, label='Actual')
plt.plot(test.index, rf_predictions, color='green', label='RandomForest Forecast')
plt.title('RandomForest Forecast vs Actual')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```
